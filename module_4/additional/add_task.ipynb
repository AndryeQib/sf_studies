```python
import pandas as pd
```

```python
data = pd.read_csv('DataSet.csv')
```

```python
data['costs'] = data['costs'].apply(lambda x: round(x))
data['income'] = data['income'].apply(lambda x: round(x))
```

```python
data_bel = data[data.arrival_airport == 'Belgorod']
```

```python
data_bel[(data_bel.income < data_bel.income.mean()) & (data_bel.tickets_count < data_bel.tickets_count.mean())]
```
